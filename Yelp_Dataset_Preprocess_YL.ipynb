## Dataset Introduction

Yelp Dataset Challenge
download data from:https://www.yelp.com/dataset/challenge

### The Challenge Dataset:
4.1M reviews and 947K tips by 1M users for 144K businesses
1.1M business attributes, e.g., hours, parking availability, ambience.
Aggregated check-ins over time for each of the 125K businesses
200,000 pictures from the included businesses


### Cities:
U.S.: Pittsburgh, Charlotte, Urbana-Champaign, Phoenix, Las Vegas, Madison, Cleveland,etc

### Files:
yelp_academic_dataset_business.json
yelp_academic_dataset_checkin.json
yelp_academic_dataset_review.json
yelp_academic_dataset_tip.json
yelp_academic_dataset_user.json


### Notes on the Dataset
Each file is composed of a single object type, one json-object per-line.
Take a look at some examples to get you started: https://github.com/Yelp/dataset-examples.



```python
import json
import pandas as pd
```

```python
file_business, file_checkin,file_reviews,file_tip,file_user=[
    'yelp_academic_dataset_business.json',
    'yelp_academic_dataset_checkin.json',
    'yelp_academic_dataset_review.json',
    'yelp_academic_dataset_tip.json',
    'yelp_academic_dataset_user.json'
]
```

```python
with open(file_business,encoding="utf8")as f:
    df_business=pd.DataFrame(json.loads(line) for line in f)
```

```python
df_business.head(5)
```

```python
with open(file_checkin,encoding="utf8")as f:
    df_checkin=pd.DataFrame(json.loads(line) for line in f)
```

```python
df_checkin.head(5)
```

```python
with open(file_reviews,encoding="utf8")as f:
    df_reviews=pd.DataFrame(json.loads(line) for line in f)
```

```python
df_reviews.head(5)
```

```python
df_business.isnull().sum()
```

```python
selected_col=[u'business_id',u'categories',u'city',u'name',u'review_count',u'stars']
```

```python
df_business_se=df_business[selected_col].dropna()
df_business_se.info()
```

```python
df_LV=df_business_se[(df_business_se['city']=='Las Vegas')]
df_LV.info()
```

```python
len(df_LV['business_id'].unique())
```

```python
res_list=[df_LV['business_id'].index[i] for i, element in enumerate(df_LV['categories']) if 'Restaurants' in element]
len(res_list)
```

```python
df_business_selected=df_LV.loc[res_list]
```

```python
df_business_selected.head(10)
```

```python
df_left=df_business_selected.set_index(['business_id'])
```

```python
df_right=df_reviews.set_index(['business_id'])
```

```python
df_left=df_left.rename(columns={'stars':'avg_stars'})
```

```python
df_left.head(5)
```

```python
df_right.head(5)
```

```python
df_joint=df_left.join(df_right)
```

```python
df_joint.head(10)
```

```python
df_joint=df_joint.reset_index()
```

```python
df_joint.head(10)
```

```python
df_joint['date']=pd.to_datetime(df_joint['date'])
```

```python
df_joint_recent=df_joint[df_joint['date']>pd.to_datetime('2016-01-01')]
```

```python
df_joint_recent.head(10)
```
